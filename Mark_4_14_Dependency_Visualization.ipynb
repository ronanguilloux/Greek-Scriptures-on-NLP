# Visualizing Grammatical Dependencies in Mark 4:14 🌱

This short notebook demonstrates how to use the [spaCy](https://spacy.io/) framework along with its built-in `displaCy` visualizer to illustrate grammatical dependencies in Ancient Greek.

We are going to use the `grc_odycy_joint_trf` transformer model to analyze Mark 4:14:
> ὁ σπείρων τὸν λόγον σπείρει.

*(The farmer sows the word.)*

```python
import spacy
from spacy import displacy

# Load the Ancient Greek transformer model
print("Loading odyCy model...")
nlp = spacy.load("grc_odycy_joint_trf")
```

## Parsing the Text

Let's take the Greek text of Mark 4:14 ('The farmer sows the word') and process it through our NLP pipeline. 

This step automatically tokenizes the words, assigns parts of speech, determines morphology, and maps out the syntactic dependencies (how the words relate to one another).

```python
# The Greek text for Mark 4:14 – 'The farmer sows the word'
verse_text = "ὁ σπείρων τὸν λόγον σπείρει."

# avoiding a little warning in nlp()
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="thinc.shims.pytorch")

# Process the text
doc = nlp(verse_text)
print(f"Successfully parsed {len(doc)} tokens.")
```

## Visualization with displaCy

Now we can use `displaCy` to render a beautiful dependency graph right here inside the notebook. 

The arrows indicate the syntactic relations (e.g., subject, object, modifier) pointing from the 'head' word to its 'child' (dependent) word.

```python
from IPython.display import HTML, display

# Set some custom options for displaCy to make the Greek text larger and more readable
options = {
    "compact": False,
    "bg": "#ffffff",
    "color": "#000000",
    "font": "Arial",
    "distance": 120
}

# Get raw HTML from displaCy (bypassing the internal buggy Jupyter import)
html_output = displacy.render(doc, style="dep", jupyter=False, options=options)

# Display it manually using modern IPython imports
display(HTML(f'<span class="tex2jax_ignore">{html_output}</span>'))
```
