# Mark on spaCy

Processing Ancient Greek text with the OdyCy transformer model.

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="thinc.shims.pytorch")

import spacy
import torch
```

```python
# Load the model
nlp = spacy.load("grc_odycy_joint_trf")

# Parse some test text
doc = nlp("Ἐν ἀρχῇ ἦν ὁ λόγος")

# Print tokens, lemmas, and parts of speech
for t in doc:
    print(f"{t.text:10} | {t.lemma_:10} | {t.pos_}")
```
